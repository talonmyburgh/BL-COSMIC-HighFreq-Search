```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make a plot showing previous SETI searches
extents = [(1e3, 10e3), (1e3, 11e3), (22230, 22240), (1419, 1421), (24e3, 50e3)]
names = ["ATA Searches", "GBT Searches", "Water Line Searches", "H Alpha Searches", "This Work (COSMIC)"]

h_scale = 1
fig, ax = plt.subplots(figsize=(12, 8))
for i, extent in enumerate(extents):
    name = names[i]
    low = extent[0]
    high = extent[1]
    ax.plot([low, high], [i * h_scale]*2, marker='o', label=name, linewidth=10, markersize=20)

bottom_lim = 0
height = (len(extents) - 1) * h_scale
plt.title("Rough Sketch of Previous SETI Search Ranges", fontsize=20)
plt.yticks(range(len(names)), names)
plt.xlabel("Frequency (MHz)", fontsize=20)
plt.xlim([bottom_lim, 55e3])
ax.tick_params(axis='both', which='major', labelsize=20)
# plt.xscale('log')
# plt.legend()
plt.show()
```
